```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
df = pd.read_csv("waste.csv", index_col=["Year", "Country"], parse_dates=True)
```

```python
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
df = df.pivot_table(values="Value", index=["Year", "COU"], columns=["Variable"])
```

```python
df.columns
```

```python
df.info()
```

```python
df = df[['Composting', 'Landfill', 'Municipal waste generated', 'Municipal waste generated per capita', 
         'Municipal waste treated', 'Recycling', 'Total Incineration',]]
```

```python
df.info()
```

```python
# Review what years lack information to drop them
df.count(level="Year")
```

```python
df = df.loc[(slice("1995", "2017"), slice(None)),:]
df.head()
```

```python
df.info()
```

```python
df.sort_index(level=1, inplace=True)
df.head(10)
```

```python
df.index.levels[1]
```

```python
# Interpolating the missing values by country to fill many nans

for country in df.index.levels[1]:
    df.loc[(slice(None), country), :] = df.loc[(slice(None), country), :].interpolate(limit_direction="both")
```

```python
# Completing the Municipal waste column by adding the parts of the same
df["Municipal waste generated"].fillna(df["Composting"] + df["Landfill"] + df["Recycling"] + df["Total Incineration"], inplace=True)
```

```python
# Droping the rows with nan values in the Recycling column as it is very difficult to estimate
df = df.dropna(subset=["Recycling"])
```

```python
df.info()
```

```python
# Function to retrieve the total population of countries with no values on the waste per capita column
def pop_per_capita(data):
    country = str(data[1])
    year = str(data[0].year)
    pop = requests.get("http://api.worldbank.org/v2/countries/" + country + "/indicators/SP.POP.TOTL?date=" + year + "&format=json")
    pop_value = pop.json()[1][0]["value"]
    return pop_value
```

```python
waste_per_capita_null = df[df["Municipal waste generated per capita"].isnull()]
```

```python
population_null = pd.Series(data=waste_per_capita_null.index.map(pop_per_capita), index=waste_per_capita_null.index)
```

```python
# Calculating the waste per capita on the missing rows
df["Municipal waste generated per capita"].fillna(1000 / population_null, inplace=True)
```

```python
# Filling the rest of the values with 0
df.fillna(0, inplace=True)
```
